```
import pandas as pd
import numpy as np

train_data=pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/titanic_data.csv")
train_data1=pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/titanic_data.csv")

train_data.info()

#we can see that Age and Cabinhave less than 891, which means they have missing values
```

```
train_data.head(3)
```

```
#dropping columns that are not useful

cols=['Name','Ticket','Cabin']
train_data=train_data.drop(cols,axis=1)
train_data1=train_data1.drop(cols,axis=1)

#axis=1 I think means drop columns

train_data.info()
```

```
#creating Dummy Variables now we convert Pclass, Sex, Embarked to columns in pandas and drop them after conversion

dummies=[]

cols=['Pclass','Sex','Embarked']

for col in cols:
  dummies.append(pd.get_dummies(train_data1[col]))
titanic_dummies=pd.concat(dummies,axis=1)

titanic_dummies
```

```
train_data1=pd.concat((train_data1,titanic_dummies),axis=1)
train_data1
```

```
train_data1=train_data1.drop(['Pclass','Sex','Embarked'],axis=1)
train_data1
```

```
#since Age has lost a lot of values, we need to replace all the missing NaNs
#which can be done with median or interpolate() function
#this function changes all the missing values with interpolate values (i.e. median)

train_data1["Age"]=train_data1["Age"].interpolate()

train_data1.info()
```

```
#Converting the dataframe to numpy array


#This way we preparethe data for ML models

x=train_data1.values

y=train_data1['Survived'].values


#x has still Survived values in it, which sould not be there
#that is why we drop in numpy column which is the1st column

x=np.delete(x,1,axis=1)
```

```
#Dividing data set into training set and test set


from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.3)

#with this step data-preprocessing is finished
```
